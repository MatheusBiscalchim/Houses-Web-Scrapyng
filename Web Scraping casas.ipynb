```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import re
```

```python
df = ['Bairro','Area','Quartos','Banheiros','Garagem','Preço']
df = pd.DataFrame(columns=df)
```

```python
def coletando_infos_casas(preco_inicial, preco_final, paginas, cidade, estado):
    service = Service()

    options = webdriver.ChromeOptions()

    driver = webdriver.Chrome(service=service, options=options)
    url = f'https://www.chavesnamao.com.br/casas-a-venda/{estado}-{cidade}/?&filtro={{"pmin":{preco_inicial},"pmax":{preco_final}}}&pg={paginas}'

    driver.get(url)
    
    casas = driver.find_elements(By.CSS_SELECTOR, 'span.contentCard')
    for i in range(len(casas)):
        casas_1.append(casas[i].text.split('\n'))

    wait = WebDriverWait(driver, 10)
    lista_final.append(casas_1)
    driver.quit()
    return lista_final
```

```python
precos = [0, 300000, 450000, 600000, 750000, 900000, 1200000]
paginas = range(5)
cidades = ['florianopolis','porto-alegre','curitiba', 'rio-de-janeiro','sao-paulo','belo-horizonte','fortaleza', 'manaus','savaldor', 'recife', 'natal', 'goiania', 'teresina', 'belem']
estados = ['sc','rs','pr', 'rj', 'sp', 'mg','ce','am','ba', 'pe', 'rn', 'go', 'pi', 'pa']
lista_final = []
casas_1 = []
for c, e in zip(cidades, estados):
    for i in (range(len(precos)-1)):
        for j in paginas:
            coletando_infos_casas(precos[i],precos[i+1],paginas[j],c,e)
```

```python
for i in range(len(lista_final)):
    print("Comecei I", i)
    for j in range(len(lista_final[i])):
        if len(lista_final[i][j]) == 12:
            new_row = {'Bairro': lista_final[i][j][2], 'Preço': lista_final[i][j][3], 'Area':lista_final[i][j][4], 'Quartos': lista_final[i][j][5], 'Banheiros': lista_final[i][j][7], 'Garagem': lista_final[i][j][9]}
            df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)
            df = df.drop_duplicates()
```
